# Text download and save function

```python
# First fill out you local path to store the dataset:

dataset_folder_path = '' # FILL OUT FIRST!
```

```python
def download_and_save_speech(links):
    
    for link in links:  
        # Get page text
        speech_page_url = link
        page = requests.get(speech_page_url).text
        soup = BeautifulSoup(page, "lxml")

        # Process page title and create appropriately named file
        title = ""
        for letter in reversed(link): # "reversed" function reverses a list
            title = title + letter
            if letter == '/':
                break
        title = title[::-1] # Reverse list
        print("Title: ", title)

        # Create a file with appropriate title
        f = io.open(dataset_folder_path + title + '.txt', "x", encoding="utf-8")

        # Save speech text to the file
        for p in soup.find(class_='transcript-inner').find_all('p'):
            f.write(p.text)
        
        print("The file:", title, "has been saved.")
        f.close()
```

```python
download_and_save_speech(urls)
```
